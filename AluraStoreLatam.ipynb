### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()
```

```python
print("Ingreso Tienda 1:", ingreso_tienda1)
print("Ingreso Tienda 2:", ingreso_tienda2)
print("Ingreso Tienda 3:", ingreso_tienda3)
print("Ingreso Tienda 4:", ingreso_tienda4)
```

# 2. Ventas por categoría

```python
categorias_tienda1 = tienda.groupby("Categoría del Producto").size()
categorias_tienda2 = tienda2.groupby("Categoría del Producto").size()
categorias_tienda3 = tienda3.groupby("Categoría del Producto").size()
categorias_tienda4 = tienda4.groupby("Categoría del Producto").size()
```

```python
print("Tienda 1")
print(categorias_tienda1)

print("\nTienda 2")
print(categorias_tienda2)

print("\nTienda 3")
print(categorias_tienda3)

print("\nTienda 4")
print(categorias_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion1 = tienda["Calificación"].mean()
calificacion2 = tienda2["Calificación"].mean()
calificacion3 = tienda3["Calificación"].mean()
calificacion4 = tienda4["Calificación"].mean()
```

```python
print("Promedio Tienda 1:", calificacion1)
print("Promedio Tienda 2:", calificacion2)
print("Promedio Tienda 3:", calificacion3)
print("Promedio Tienda 4:", calificacion4)
```

# 4. Productos más y menos vendidos

```python
ventas_productos1 = tienda["Producto"].value_counts()
ventas_productos2 = tienda2["Producto"].value_counts()
ventas_productos3 = tienda3["Producto"].value_counts()
ventas_productos4 = tienda4["Producto"].value_counts()
```

```python
print("Más vendido Tienda 1:")
print(ventas_productos1.head(1))

print("Menos vendido Tienda 1:")
print(ventas_productos1.tail(1))
```

# 5. Envío promedio por tienda

```python
envio1 = tienda["Costo de envío"].mean()
envio2 = tienda2["Costo de envío"].mean()
envio3 = tienda3["Costo de envío"].mean()
envio4 = tienda4["Costo de envío"].mean()
```

```python
print("Costo envío promedio tienda 1:", envio1)
print("Costo envío promedio tienda 2:", envio2)
print("Costo envío promedio tienda 3:", envio3)
print("Costo envío promedio tienda 4:", envio4)
```

```python
import matplotlib.pyplot as plt
```

Gráfica Ingresos por tienda

```python
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.bar(tiendas, ingresos)

plt.title("Ingresos totales por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")

plt.show()
```

Gráfica Productos más vendidos

```python
top_productos = tienda["Producto"].value_counts().head(5)

plt.bar(top_productos.index, top_productos.values)

plt.title("Top 5 productos más vendidos - Tienda 1")
plt.xlabel("Producto")
plt.ylabel("Cantidad vendida")

plt.xticks(rotation=45)

plt.show()
```

Gráfica Calificaciones de clientes

```python
calificaciones = [calificacion1, calificacion2, calificacion3, calificacion4]

plt.plot(tiendas, calificaciones, marker='o')

plt.title("Calificación promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación promedio")

plt.show()
```

**Introducción**

El objetivo de este análisis fue evaluar el desempeño de las cuatro tiendas de la cadena Alura Store para ayudar al Sr. Juan a decidir cuál tienda debería vender para iniciar un nuevo emprendimiento. Para ello se analizaron diferentes aspectos como los ingresos totales, las categorías de productos más vendidas, las calificaciones de los clientes, los productos más y menos vendidos y el costo promedio de envío.

**Desarrollo del análisis**

A partir del análisis de los datos y de los gráficos generados se pudieron identificar diferencias importantes entre las cuatro tiendas.

En primer lugar, se analizaron los ingresos totales, donde se observó que algunas tiendas generan mayores ventas que otras. Esto permite identificar cuáles tienen un mejor rendimiento económico.

También se analizaron las categorías de productos, lo que permitió identificar cuáles productos tienen mayor demanda entre los clientes y cuáles se venden menos.

Otro aspecto importante fue la calificación promedio de los clientes, ya que permite medir el nivel de satisfacción con los productos y el servicio de cada tienda.

Además, se identificaron los productos más y menos vendidos, lo que ayuda a entender qué artículos tienen mayor aceptación en el mercado.

Finalmente, se analizó el costo promedio de envío, un factor que puede influir en la decisión de compra de los clientes.

**Conclusión**

Con base en los resultados obtenidos, se concluye que la Tienda 4 presenta el menor rendimiento en comparación con las demás tiendas, ya que genera menores ingresos y no destaca en los demás indicadores analizados.

Por esta razón, se recomienda que el Sr. Juan venda la Tienda 4, ya que es la menos eficiente dentro de la cadena Alura Store. Esta decisión permitiría enfocar los recursos en las tiendas con mejor desempeño y mayor potencial de crecimiento.
